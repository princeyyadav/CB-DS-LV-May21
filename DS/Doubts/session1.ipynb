```python
import numpy as np
import matplotlib.pyplot as plt
from mysklearn.NNeighbours import KNN
from mysklearn.linearmodels import LogisticRegression
```

```python
def predict(X, y, test_point, k=3):
        # compute distance
        distance = ((X - test_point)**2).sum(axis=1)**0.5
        print("distance computed for all training points", distance.shape)
        knn_indexes = np.argsort(distance)[:k] # indices of k nearest neighbour
        print(f"Index of {k} nearest Neighbour {knn_indexes}")
        # get the category of my NN
        knn_cat = y[knn_indexes]
        print(f"Category ofnearest neighbour {knn_cat}")
        cls, count = np.unique(knn_cat, return_counts=True)
        print(cls, count)
        d = {key:v for key, v in zip(cls, count)}
        print("No. of samples of each category: ", d)

        max_count_idx = np.argmax(count)
        print("Index of the class for which we have maximum no. of samples: ", max_count_idx)
        pred_cat = cls[max_count_idx] # category with maximum count
        print("Predicted category of test point is:", pred_cat)
        return pred_cat
```

```python
## Get data
data_path = "E:/CB-DS-LV-May21/DS/data/knn/classification/"
X = np.load(data_path+"X_data.npy")
y = np.load(data_path+"Y_data.npy").reshape(-1,1)
print(X.shape, y.shape)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
# dry run for one test point
tp = [2, 6]
print(predict(X, y, tp, k=10))
```
