## Linear Regression
### Assumption: 
Linear relationship between input variable and output variable. Meaning for a unit change in x, there is a constant change in y
### Aim
To find a line that fits the data best. Best fit means line for which the error, i.e, the difference between true and predicted value is minimized for all data points. Finding best fit line means finding the optimal m and c for which loss function is the minimum.    
### Predicted value 
$\hat{y} = m*x + c$
### Loss/ Cost Function
## $\sum_{i}(y_{i} - \hat{y}_{i})^{2}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.load(r"..\data\knn\regression\X_data.npy")
y = np.load(r"..\data\knn\regression\y_data.npy")
```

```python
print(X.shape) # 50 samples, 1 feature
print(y.shape)

print(X[0], y[0])
```

```python
# visualize
plt.scatter(X, y)
plt.xlabel("Feature")
plt.ylabel("Output (y)")
plt.title("Output vs feature")
plt.show()
```

```python
# optimal m and c: value of m and c for which the loss function/ Cost function, J(m,c) is minimum.

x_bar = X.mean(0)
y_bar = y.mean(0)
xy_bar = (X[:,0]*y).mean()
x2_bar = (X**2).mean(0)

m = (xy_bar - y_bar*x_bar)/(x2_bar + x_bar**2)
c = y_bar - m*x_bar

print("Optimal value of parameters")
print("="*30)
print(f"m: {m}, c: {c}")
```

```python
# x = np.linspace(X[:,0].min()-1, X[:,0].max()+1, 50) # test_points
x = X[:,0]
ypred = m*x + c # prediction

plt.scatter(X, y, label="Training data")
plt.plot(x, ypred, c="red", label="Prediction")
plt.legend()
plt.show()
```

```python
test_point = 0.5
yt = m*test_point + c
plt.scatter(X, y, label="training data")
plt.plot(x, ypred, c="crimson", linewidth=1, label="Prediction")
plt.scatter(test_point, yt, c="orange", label="test point")
plt.legend()
plt.show()
```

### How the regression data was generated
1. Generate some points for x  
2. Assume some m and c  
3. y = m.x + c  
4. Add some scaled gaussian noise to y  

```python
noise = np.random.randn(y.shape[0])
y_ = ypred + noise*10
plt.scatter(x, y_)
plt.show()
```
